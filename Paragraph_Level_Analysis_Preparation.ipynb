```python
import csv
import re

min_char_count = 50

def to_paragraphs(file_name, to_csv):
    print(f" \n\n\nExtracting {file_name}:")
    # Load and clean the text
    with open(file_name, 'r', encoding='utf-8') as file:
        raw_text = file.read()

    # Split text into raw paragraphs based on double newlines or formatting marks
    raw_paragraphs = raw_text.split('\n\n')
    # Pattern that allows letters, single spaces, and single basic punctuation
    allowed_pattern = re.compile(r"[^A-Za-z\s.,?!:;\'\"()\-\n]")
    # Pattern to remove long sequences of dashes or periods (like ------ or ....)
    repeated_punct_pattern = re.compile(r"[-]{2,}|[.]{2,}")


    # Process and clean paragraphs
    paragraphs = []
    i = 1
    for para in raw_paragraphs:
        cleaned = ' '.join(para.split())  # normalize whitespace
        cleaned = re.sub(r'(\w)-\s(\w)', r'\1\2', cleaned)  # join broken words
        cleaned = re.sub(r'\s{2,}', ' ', cleaned)  # replace multiple spaces with one
        cleaned = repeated_punct_pattern.sub('', cleaned)  # remove repeated dashes or dots
        cleaned = allowed_pattern.sub('', cleaned)  # remove unwanted characters

        if len(cleaned) >= min_char_count:
            new_add = cleaned.strip()
            paragraphs.append(new_add)
            print(i, '[', len(new_add), ']', new_add)
            i += 1

    if to_csv:
        # Save to CSV
        output_path = './[p]' + file_name + '.csv'
        with open(output_path, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['text'])
            for p in paragraphs:
                writer.writerow([p])

        print(f"Saved {len(paragraphs)} cleaned paragraphs to {output_path}")
```

```python
#to_paragraphs('20110629_Lithuania_National Cybersecurity Strategy_ENG.txt', True)
to_paragraphs('20030201_USA_National Cybersecurity Strategy_ENG.txt', True)
```
